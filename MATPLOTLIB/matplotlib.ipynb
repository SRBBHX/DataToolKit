```python
x=np.random.rand(50)
y1=np.random.rand(50)
y2=np.random.rand(50)
```

```python
plt.plot(x,y1, color="red", linestyle="--",marker="o",markersize=3,linewidth=.5,label="x with y1")
plt.plot(x,y2,color="b", marker="o", linestyle=":",markersize=3,linewidth=.5,label="x with y2")
plt.title("x with y1 and y2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
